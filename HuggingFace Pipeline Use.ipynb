```python
from transformers import pipeline

# 1. 모델 로드 (의료 요약 전용 모델)
summarizer = pipeline("summarization", model="Falconsai/medical_summarization")

# 2. 의료 텍스트 샘플 (영문 의료 데이터가 더 정확도가 높습니다)
medical_text = """
The patient is a 65-year-old male with a history of hypertension and type 2 diabetes. 
He presented with chest pain and shortness of breath. 
An EKG was performed, showing ST-segment elevation. 
The patient was immediately taken to the cardiac catheterization lab.
"""

# 3. 결과 출력
summary = summarizer(medical_text, max_length=50, min_length=10, do_sample=False)
print("--- Medical Summary ---")
print(summary[0]['summary_text'])
```

```python
import streamlit as st
from transformers import pipeline

st.title("AI Medical Symptom Checker")

@st.cache_resource
def load_model():
    return pipeline("text2text-generation", model="google/flan-t5-small")

generator = load_model()

symptom = st.text_input("Enter your symptom (example: headache, fever)")

if st.button("Analyze Symptom"):
    if symptom == "":
        st.warning("Please enter a symptom")
    else:
        prompt = f"""
        A patient has this symptom: {symptom}.
        List 5 possible diseases.
        """
        result = generator(prompt, max_new_tokens=80)
        st.subheader("Possible illnesses")
        st.write(result[0]["generated_text"])
        st.info("This AI is for educational purposes only.")
```

```python
from transformers import pipeline

# 1. 모델 로드 (처음 한 번만 실행)
summarizer = pipeline("summarization", model="Falconsai/medical_summarization")

# 2. 테스트용 의료 텍스트
medical_text = """
The patient is a 65-year-old male with a history of type 2 diabetes and hypertension. 
He presents with a sudden onset of chest pain and shortness of breath. 
An EKG was performed, showing ST-segment elevation. 
The medical team administered aspirin and prepared the patient for emergency angioplasty.
"""

# 3. 요약 실행
result = summarizer(medical_text, max_length=50, min_length=10, do_sample=False)

# 4. 결과 출력 (이 부분이 화면에 보여야 합니다!)
print("-" * 30)
print("[원본 의료 텍스트]")
print(medical_text.strip())
print("-" * 30)
print("[AI 요약 결과]")
print(result[0]['summary_text'])
print("-" * 30)
```
